```python
#Task 1
class PlayerEarning:
    def __init__(self,name):
        self.name = name
    def calculateTotal(self,earning,goal=0):
        self.earning = earning
        self.goal = goal
        self.bonus = 0
        if self.goal > 30:
            self.bonus = (5/100) * self.earning + 10000
        elif 0 < self.goal <= 30:
            self.bonus = (5/100) * self.earning
        else:
            self.bonus = 0
    def printDetails(self):
        print(f"Player Name: {self.name}\nPlayer Season Earning without bonus: {self.earning}\nBonus: {self.bonus}\nPlayer Season Earning After Bonus: {self.earning+self.bonus}")

       
print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
#Task 11
class Transport:
    total_traveller = 0
    
    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare
    
    def __str__(self):
        s =  "Name: "+self.name+", Base fare: "+str(self.baseFare)
        return s

class Bus(Transport):
    def __init__(self,name,fare):
        super().__init__(name,fare)
        self.output = ""
        print(f"Base-fare of {self.name} is {self.baseFare} Taka ")
    def addPassengerWithBags(self,*args):
        self.info = args
        Transport.total_traveller += int((len(self.info)/2))
    def __str__(self):
        for i in range(len(self.info)):
            if (i%2==0):
                self.output+= f"Name: {self.info[i]}"
            else:
                if (self.info[i]<=2):
                    newFare = self.baseFare
                elif (2<self.info[i]<=5):
                    newFare = self.baseFare+60
                elif (self.info[i]>5):
                    newFare = self.baseFare+105
                self.output+=f", Fare: {str(newFare)}\n"
        return f"{super().__str__()}\nTotal Passenger(s): {str(Transport.total_traveller)}\nPassenger details:\n{self.output}"

class Train(Transport):
    def __init__(self,name,fare):
        super().__init__(name,fare)
        self.output = ""
        print(f"Base-fare of {self.name} is {self.baseFare} Taka ")
    def addPassengerWithBags(self,*args):
        self.info = args
        Transport.total_traveller += int((len(self.info)/2))
    def __str__(self):
        for i in range(len(self.info)):
            if (i%2==0):
                self.output+= f"Name: {self.info[i]}"
            else:
                if (self.info[i]<=2):
                    newFare = self.baseFare
                elif (2<self.info[i]<=5):
                    newFare = self.baseFare+60
                elif (self.info[i]>5):
                    newFare = self.baseFare+105
                self.output+=f", Fare: {str(newFare)}\n"
        return f"{super().__str__()}\nTotal Passenger(s): {str(Transport.total_traveller)}\nPassenger details:\n{self.output}"



t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6,  "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller)
```
