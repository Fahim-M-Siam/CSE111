```python
#Task 12
class TaxiLagbe:
  def __init__(self,tn,lo):
    self.tn = tn
    self.lo = lo
    self.l = []
    self.pfl = []

  def addPassenger(self,*args):
    self.pnl = []
    for i in args:
      si = i.split('_')
      self.pnl.append(si[0])
      self.l.append(si[0])
      self.pfl.append(si[1])
      
    if len(self.l)>4:
      print('Taxi Full! No more passengers can be added.')
      self.l.pop()
      self.pfl.pop()    
    else:
      for j in self.pnl:
        print(f'Dear {j}! Welcome to TaxiLagbe.')

  def printDetails(self):
    print('Trip info for Taxi number:',self.tn)
    print(f'This taxi can cover only {self.lo} area.')
    print('Total passengers:',len(self.l))
    print('Passenger lists:\n',','.join(self.l))
    tcf = 0
    for i in self.pfl:
      tcf += int(i)
    print(f"Total collected fare: {tcf} Taka")
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#Task 14
class StudentDatabase:
  def __init__(self,name,id):
    self.name = name
    self.id = id
    self.grades = {}
    self.cl = []
    self.ctt = None
  
  def calculateGPA(self,cacgl,ctt):
    gpal = []
    self.ctt = ctt
    for i in cacgl:
      x = i.split(':')
      self.cl.append(x[0])
      gpal.append(x[1])
    sum = 0
    for j in gpal:
      sum += float(j)*3
    gpa = float(sum/(len(gpal)*3))
    d = {}
    d[tuple(self.cl)] = gpa
    self.grades[ctt] = d

  def printDetails(self):
    print('Name:',self.name)
    print('ID:',self.id)
    for k,v in self.grades.items():
      for m,n in v.items():
        print(f'Courses taken in {k}:')
        for x in m:
          print(x)
        print('GPA:', n)
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'],
'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'],
'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#Task 20
class Student:
  def __init__(self, n=None, d=None):
    self.name = n
    self.dept = d
    self.lst = []
    if self.name == None and self.dept == None:
      print("Student name and department need to be set")
    elif self.name != None and self.dept == None:
      print(f"Department for {self.name} needs to be set")
    else:
      print(f"{self.name} is from {self.dept} department")

  def update_name(self,n):
    self.name = n

  def update_department(self,d):
    self.dept = d

  def enroll(self, *args):
    self.course = list(args)
    for i in self.course:
      if i not in self.lst:
        self.lst.append(i)
      else:
        self.lst = i

  def printDetail(self):
    print(f"Name: {self.name} \nDepartment: {self.dept} \n{self.name} enrolled in {len(self.course)} course(s):")
    print(", ".join(self.lst))



s1 = Student()
print("=========================")
s2 = Student("Carol")
print("=========================")
s3 = Student("Jon", "EEE")
print("=========================")
s1.update_name("Bob")
s1.update_department("CSE")
s2.update_department("BBA")
s1.enroll("CSE110", "MAT110", "ENG091")
s2.enroll("BUS101")
s3.enroll("MAT110", "PHY111")
print("###########################")
s1.printDetail()
print("=========================")
s2.printDetail()
print("=========================")
s3.printDetail()
```
